```python
import cv2 # for open cv
```

```python
import matplotlib.pyplot as plt
```

```python
config_file = 'ssd_mobilenet_v3_large_coco_2020_01_14.pbtxt'
frozen_model = 'frozen_inference_graph.pb'
```

```python
model = cv2.dnn_DetectionModel(frozen_model,config_file)
```

```python
classLabels = []
file_name = 'Labels.txt'
with open(file_name,'rt') as fpt:
    classLabels = fpt.read().rstrip('\n').split('\n')
#     classLabels.append(fpt.read())
    #adding all the Labels to class labels
```

```python
print(classLabels)
```

```python
print(len(classLabels))
```

Read a image 

```python
model.setInputSize(320,320)
model.setInputScale(1.0/127.5) ##255/2 =127.5
model.setInputMean((127.5,127.5,127.5)) ## mobilenet
model.setInputSwapRB(True)
```

```python
img = cv2.imread('hi.jpg')
```

```python
plt.imshow(img)
```

```python
plt.imshow(cv2.cvtColor(img , cv2.COLOR_BGR2RGB))
```

```python
ClassIndex , confidece ,bbox = model.detect(img,confThreshold=0.5)
```

```python
print(ClassIndex)
```

```python
font_scale = 3
font = cv2.FONT_HERSHEY_PLAIN
for ClassInd , conf , boxes in zip(ClassIndex.flatten(),confidece.flatten(),bbox): #adding boxes to images
    cv2.rectangle(img,boxes,(255,0,0),2)
    cv2.putText(img,classLabels[ClassInd-2],(boxes[0]+10,boxes[1]+40), font , fontScale=font_scale,color=(0,255,0),thickness=3)
```
